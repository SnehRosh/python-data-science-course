```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(data=iris, x='sepal_length',
                y='sepal_width',
                hue='species',
            )
plt.grid()
```

```python
x=['petal_length']
plt.hist(x, bins=list(range(6)),color='red')
plt.show()
```

```python
x=iris[['sepal_width','sepal_length']]
plt.boxplot(x)
plt.show()
```

```python
sns.lineplot(data=iris,
             x='sepal_length',
             y='petal_length',
             hue='species'
            )
plt.legend()
plt.show()
```
